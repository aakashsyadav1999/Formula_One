```python
car_per_win = race_details.groupby(['Car', 'Winner'])['Winner'].count().reset_index(name='Wins')
car_per_win = car_per_win.sort_values(by='Wins', ascending=True)
fig = px.bar(car_per_win, x='Car', y='Wins', color='Winner')
fig.update_layout(title='Number of Wins per Car', xaxis_title='Car', yaxis_title='Number of Wins')
fig.show()
```

```python
df_lap_times = race_details[['Laps', 'year']]
df_lap_times = df_lap_times.dropna()
df_lap_times['Laps'] = df_lap_times['Laps'] * 60
g = sns.FacetGrid(df_lap_times, col='year', col_wrap=4, height=3, margin_titles=True)
g.map(plt.hist, 'Laps', bins=20)
avg_lap_times = df_lap_times.groupby('year').mean()
plt.figure()
plt.plot(avg_lap_times.index, avg_lap_times['Laps'], marker='o')
plt.xlabel('Year')
plt.ylabel('Average Lap Time (seconds)')
plt.show()
```

```python
race_details['Time'] = pd.to_timedelta(race_details['Time'], errors='coerce')
df_race = race_details.dropna(subset=['Time'])
df_race['LapTime'] = df_race['Time'].dt.total_seconds() / df_race['Laps']
df_avg_lap_time = df_race.groupby(['Winner', 'Car', 'year'], as_index=False)['LapTime'].mean()
fig = px.line(df_avg_lap_time, x='year', y='LapTime', color='Winner', hover_data=['Car'])
fig.update_layout(
    title='Average Lap Time by Driver, Car, and Year',
    xaxis_title='Year',
    yaxis_title='Average Lap Time (seconds)',
    hoverlabel=dict(font_size=12, font_family='Arial')
)

fig.show()
```

```python
df_race['Date'] = pd.to_datetime(df_race['Date'])
df_race['Year'] = df_race['Date'].dt.year
laps_per_year = df_race.groupby('Year')['Laps'].mean()
plt.plot(laps_per_year.index, laps_per_year.values)
plt.title('Distribution of Laps per Race over the Years')
plt.xlabel('Year')
plt.ylabel('Average Laps per Race')
plt.show()
```

We can see in this how laps have been dropped as compared to past

Analysis of fastest laps 

```python
df_race['Time'] = df_race['Time'].astype(str)
df_race['Time'] = pd.to_timedelta(df_race['Time'])
df_race['Lap Time'] = df_race['Time'] / df_race['Laps']
fastest_lap = df_race.groupby('Grand Prix')['Lap Time'].min()
df_fastest_lap = pd.merge(df_race, fastest_lap, on=['Grand Prix', 'Lap Time'])
df_fastest_lap = df_fastest_lap[['Grand Prix', 'Date', 'Winner', 'Car', 'Laps', 'Lap Time']].drop_duplicates()
df_fastest_lap = df_fastest_lap.sort_values(by='Lap Time')
print(df_fastest_lap)
```

```python
fig = px.line(df_fastest_lap, x='Grand Prix', y='Lap Time', color='Winner', 
              hover_data=['Car', 'Date'], title='Fastest Lap Times by Grand Prix')
fig.update_traces(mode='markers+lines', marker=dict(size=10))
fig.update_layout(
    title_font=dict(size=20),
    xaxis_title='Grand Prix',
    yaxis_title='Lap Time',
    font=dict(size=14),
    coloraxis=dict(colorscale='Viridis')
)
fig.show()
```

Keke Rosberg ROS is topping this chart


```python
df_winners = df_race.groupby(['Winner', 'Car']).size().reset_index(name='Wins')
df_winners = df_winners.sort_values(by=['Wins'], ascending=False)
fig = px.scatter(df_winners, x="Wins", y="Winner", color="Car", size="Wins", hover_name="Car",
                 hover_data={"Wins": True},
                 title="Most Successful Drivers and Teams in Terms of Race Wins",
                 labels={"Wins": "Race Wins", "Winner": "Driver"})
fig.update_layout(xaxis_title="Race Wins", yaxis_title="Driver", legend_title="Team")
fig.show()
```
